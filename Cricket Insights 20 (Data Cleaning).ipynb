```python
#importing necessary libraries
import pandas as pd
import json
```

```python
with open('t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
#To check the rows and columns in the dataframe
df_match.shape
```

```python
#Renaming the scorecard to match_id column
df_match.rename({'scorecard':'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#here I am storing the team name in dictionary for both combinations so that I can have a common
#column in both tables to make it a foreign key.
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + " Vs " + row['team2']
    key2 = row['team2'] + " Vs " + row['team1']
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
    
match_ids_dict
```

**Batting Summary**

```python
with open('t20_wc_batting_summary.json') as f:
    data = json.load(f)

    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary']) #'extend' here joins the array in batting summary element. 
all_records

df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
#Finding that the player is out or not out
df_batting['out/notout'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'notout')
df_batting.head(11)
```

```python
#Dropping the dismissal column
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head(10)
```

```python
#replacing the special character with nothing
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', '')) 
df_batting.head(10)
```

```python
#Making a new column named 'match_id' by mapping the dictionary on the match column
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head(10)
#match_ids_dict["Namibia Vs Sri Lanka"]
#match_ids_dict["U.A.E. Vs Netherlands"]
```
